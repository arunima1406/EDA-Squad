# > Dropping unnecessary columns

```python
x=df
x=x.drop(labels='Name',axis=1)
x=x.drop(labels='Ticket',axis=1)
x=x.drop(labels='Cabin',axis=1)
x=x.drop(labels='Survived',axis=1)
x=x.drop(labels='PassengerId',axis=1)
x=x.drop(labels='Fare',axis=1)
x=x.drop(labels='Embarked',axis=1)
x=x.values
y=df.iloc[:,1].values
```

```python
print (x)
```

```python
print (y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct1 = ColumnTransformer(transformers=[('encoding',OneHotEncoder(),[1])],remainder='passthrough')
x=np.array(ct1.fit_transform(x))
```

# > Splitting data into training set and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
x_train = x_train.astype(float)  # Convert to floating-point
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```
